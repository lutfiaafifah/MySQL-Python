```python
join3 = pd.read_sql_query('SELECT ord.orderNumber, ord.status, det.productCode, SUM(det.priceEach) AS totalPrice \
                           FROM orders ord \
                           JOIN orderdetails det \
                           ON ord.orderNumber = det.orderNumber GROUP BY ord.orderNumber', connection)
```
